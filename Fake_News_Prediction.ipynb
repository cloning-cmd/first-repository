```python
import numpy as np
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
news_dataset=pd.read_csv('/content/archive (1).zip')
news_dataset.shape
news_dataset.head()
```

```python
news_dataset.isnull().sum()
```

```python
news_dataset = news_dataset.fillna('')
```

```python
news_dataset['content']=news_dataset['title']+' '+news_dataset['text']
```

```python
print(news_dataset['content'])
```

```python
porter_stem=PorterStemmer()
import nltk
nltk.download('stopwords')
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

def stemming(content):
    # Clean and tokenize
    content = re.sub('[^a-zA-Z]', ' ', content)
    content = content.lower()
    content = content.split()

    # Load stopwords and initialize stemmer
    stop_words = set(stopwords.words('english'))
    porter_stem = PorterStemmer()

    # Stem and remove stopwords
    stemmed_words = [porter_stem.stem(word) for word in content if word not in stop_words]

    # Join back into a string
    stemmed_content = ' '.join(stemmed_words)
    return stemmed_content
```

```python
news_dataset['content'] = news_dataset['content'].apply(stemming)
```

```python
vectorise=TfidfVectorizer()
X=vectorise.fit_transform(news_dataset['content'])
print(X)
```

```python
Y=news_dataset['label']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of the training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy of test data :' , test_data_accuracy)
```
