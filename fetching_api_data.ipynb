```python
import requests
import json
import pandas as pd

# Base URL for the PokeAPI
base_url = "https://pokeapi.co/api/v2/pokemon/"

# Function to fetch info about any Pok√©mon
def get_pokemon_info(name):
    url = base_url + name  # Build the full URL
    response = requests.get(url)

    if response.status_code == 200:
        pokemon_data = response.json()  # Convert JSON response to dict

        # Pretty-print JSON (optional, for viewing)
        print(json.dumps(pokemon_data, indent=2))

        return pokemon_data
    else:
        print("Not retrieved: ", response.status_code)
        return None

# Call the function and get data for 'ditto'
pokemon_data = get_pokemon_info("ditto")
print(pokemon_data);

```
