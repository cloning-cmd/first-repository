```python
# Import the requests library to fetch the CSV from the web
import requests

# Import pandas for working with tabular data
import pandas as pd

# Import StringIO to treat a string like a file (needed for pandas to read)
from io import StringIO

# The URL of the CSV file you want to download
url = 'https://people.sc.fsu.edu/~jburkardt/data/csv/addresses.csv'

# Send an HTTP GET request to the URL and store the response
response = requests.get(url)

# Convert the text content of the response to a file-like object
csv_data = StringIO(response.text)

# Define the column names you want for the DataFrame
column_names = ['First Name', 'Last Name', 'Address', 'City', 'State', 'ZIP']

# Read the CSV data into a pandas DataFrame
# - names=column_names assigns your custom headers
# - header=None tells pandas not to treat the first line as a header
df = pd.read_csv(csv_data, names=column_names, header=None)

# Print the first few rows of the DataFrame to verify it's loaded correctly
print(df.head())
```

```python
# Path to the TSV file
# Specify tab character as the delimiter
# Custom column names
# Don't treat the first line as header; use the provided names

# Read a TSV (Tab-Separated Values) file using pandas
df3 = pd.read_csv('sample-1.tsv',sep='\t',names=['sno', 'movie', 'release year', 'votes', 'genre'],header=None )

# Display the first few rows of the DataFrame to confirm successful import
df3.head()
```

```python
#use cols only take those colums of dataset which are needed 
df = pd.read_csv(
    'sample-1.tsv',                         # TSV file path
    sep='\t',                               # Tab-separated
    names=['sno', 'movie', 'genre'],  # Custom headers
    header=None,                            # Don't use first line as header
    usecols=[0, 1,4]                       # Select only columns by index: sno, movie, genre
)
```

```python
df = pd.read_csv(
    'sample-1.tsv',                                # Path to the TSV file
    sep='\t',                                      # Use tab as the delimiter
    names=['sno', 'movie', 'release year', 'votes', 'genre'],  # Custom column names
    header=None,                                   # Do not treat the first row as header
    encoding='utf-8',                              # Set file encoding to UTF-8
    on_bad_lines='skip'                            # Skip lines with malformed data (bad lines)
)
df.head()
```
